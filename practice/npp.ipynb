```python
import numpy as np
# 1.
a=np.arange(0,1000).reshape(20,50)
print(f"{a[0::3]}\n\n{a[:,-10:]}\n\n{a[a%7==0]}")
a[a>900]=-1
print(f"{a}\n\n{a.flatten()}")

```

```python
# 2.
A=np.random.randint(0,100,(200,5))
B=np.random.randint(-5,6,(1,5))
print(f"{A}\n\n{B}")
A_norm=(A-A.mean(axis=0))/A.std(axis=0)
print(f"\n\n{A_norm}")
A_BONUS=A_norm+B
print(f"\n\n{A_BONUS}")
mean_per_subject=A_BONUS.mean(axis=0)
std_per_subject=A_BONUS.std(axis=0)

print(f"\n\n{mean_per_subject}\n\n{std_per_subject}\n\n{A_BONUS.max(axis=0)}\n\n{A_BONUS.min(axis=0)}\n\n")

count=(A_BONUS > (mean_per_subject+1.5*std_per_subject)).sum(axis=0)
print(count)
A_final=np.clip(A_BONUS,-2,+2)
print(f"\n\n{A_final}")
```

```python
rl=np.random.rand(500,20)
print(f"{rl}\n\n")
rl_center=rl-rl.mean(axis=0)
print(f"{rl_center}\n\n")
rl_cov=np.cov(rl_center,rowvar=False)
print(rl_cov)
vals,vecs=np.linalg.eig(rl_cov)
vals.sort()
print(f"\n\n{vals}")
```
